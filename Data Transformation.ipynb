Data transformation dalam Python dapat digunakan dengan fungsi boolean, map dan cut. Pada bagian ini kita akan fokus untuk mengubah data menggunakan metode replace().

Masukkan Kode Kita

```python
import pandas as pd 
```

```python
data = pd.read_csv('RoyalCollege.csv')
```

```python
data
```

Kita ubah kategori data jenis kelamin dimana laki-laki memiliki value "0" dan Perempuan memiliki value 1. Untuk itu kita membuat sebuah dataframe baru dengan nama "jknum"

```python
data['jknum'] = (data['jk'] == 'Perempuan').astype(int)
```

```python
data
```

Seperti yang kalian bisa lihat, Data Jenis kelamin telah diubah menjadi "0" dan "1" dan dapat dianalisa apabila sudah diubah ke bentuk ini. Selanjutnya ayo kita transformasi data ordinal "Pendidikan" karena ini 3 kita menggunakan fungsi map

```python
pendidikan_map = {'Sarjana' : 1, 'Magister' : 2, 'Doktor': 3}
data['pendordinal'] = data['pendidikan'].map(pendidikan_map)
```

```python
data
```

Dengan ini data sudah dibuah sesuai kategorik yang kita inginkan. Sekarang kita beralih ke data numerik. yang pertama adalah usia. Kita ubah sesuai ketentuan sebagai berikut:
><35 tahun   = 1
>"35-60 tahun = 2"
>">60 tahun   = 3"

Kita menggunakan fungsi cut dengan bins dan labels dimana kerja bins adalah dengan memasang rentang. contoh bins = [0,10,25,30,float('inf')] berarti rentang 0-10,10-25,25-30, dan 30- infinity. lalu gunakan label untuk kategori data seperti labels = [1,2,3] kemudian buat column di dataframe baru untuk menampung hasil cut berdasarkan bins dan labels. Contohnya sebagai berikut

```python
usia_bins = [0, 35, 60, float('inf')]
usia_labels = [1,2,3]
data['usianum'] = pd.cut(data['usia'], bins=usia_bins, labels=usia_labels, right= False)
```

```python
data
```
